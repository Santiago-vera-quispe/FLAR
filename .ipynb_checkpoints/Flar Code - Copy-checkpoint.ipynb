```python
import pandas as pd

# Ruta y nombres de archivos
archivos = {
    "tasas": "tasas_de_interes.xlsx",
    "reservas": "reservas_internacionales_netas.xlsx",
    "laboral": "mercado_laboral.xlsx",
    "precios": "precios.xlsx",
    "fiscal": "balance_fiscal.xlsx"
}

# Función para cargar y procesar cada archivo
def procesar_archivo(nombre_archivo, nombre_variable):
    df = pd.read_excel(nombre_archivo)
    
    # Convertimos la columna timePeriod a datetime y la usamos como índice
    df['timePeriod'] = pd.to_datetime(df['timePeriod'])
    df.set_index('timePeriod', inplace=True)
    
    # Agrupar por país e indicador, resamplear a trimestres y sacar el promedio
    df_trimestral = df.groupby(['refAreaID', 'indicatorID']).resample('Q').mean(numeric_only=True).reset_index()
    
    # Renombramos la columna de valor observado
    df_trimestral = df_trimestral.rename(columns={'obsValue': nombre_variable})
    
    # Solo nos quedamos con las columnas necesarias
    return df_trimestral[['refAreaID', 'timePeriod', nombre_variable]]

# Procesamos todos los archivos
df_tasas = procesar_archivo(archivos['tasas'], 'Interest_Rate_it')
df_reservas = procesar_archivo(archivos['reservas'], 'Capital_Flight_it')
df_laboral = procesar_archivo(archivos['laboral'], 'Unemployment_it')
df_precios = procesar_archivo(archivos['precios'], 'Inflation_it')
df_fiscal = procesar_archivo(archivos['fiscal'], 'Fiscal_Stimulus_it')

# Combinamos todos los DataFrames en uno solo usando merge
df_final = df_tasas.merge(df_reservas, on=['refAreaID', 'timePeriod'], how='outer')
df_final = df_final.merge(df_laboral, on=['refAreaID', 'timePeriod'], how='outer')
df_final = df_final.merge(df_precios, on=['refAreaID', 'timePeriod'], how='outer')
df_final = df_final.merge(df_fiscal, on=['refAreaID', 'timePeriod'], how='outer')

# Visualizamos los primeros registros
print(df_final.head())
```

```python
import pandas as pd

# Load the existing df_final (assuming you have saved it to a file or still have it in your notebook)
# Replace 'df_final.csv' with the actual path if you saved it
df_final = pd.read_csv('df_final.csv')
# If df_final was an Excel file:
# df_final = pd.read_excel('df_final.xlsx')

# Create a DataFrame from the consumption data (assuming you have it in a dictionary or list of lists)
consumption_data = {
    'Fecha': ['2018T1', '2018T2', '2018T3', '2018T4', '2019T1', '2019T2', '2019T3', '2019T4',
              '2020T1', '2020T2', '2020T3', '2020T4', '2021T1', '2021T2', '2021T3', '2021T4',
              '2022T1', '2022T2', '2022T3', '2022T4', '2023T1', '2023T2', '2023T3', '2023T4',
              '2024T1', '2024T2', '2024T3', '2024T4'],
    'Privado Perú': [27416, 25226, 29457, 27627, 26792, 25384, 26468, 24944,
                     25523, 24516, 27980, 28425, 28134, 25635, 27233, 27305,
                     28643, 28289, 33140, 34491, 35928, 31107, 36587, 35019,
                     34176, 30689, 33892, 28326],
    'Privado Colombia': [None, None, None, None, 178886.2, 180715.1, 182043.6, 183003.1,
                         185302.9, 187730.5, 190314.2, 192408.4, 192344.6, 160751.5,
                         176263.8, 194548.1, 199532.6, 195708.3, 208684.4, 219637.7,
                         222322.5, 224702.4, 225853.8, 224948.3, 226127.8, 224967,
                         226059.5, 225781.7]
}
df_consumption = pd.DataFrame(consumption_data)

# Function to map the 'Fecha' format to a consistent 'timePeriod' and 'refAreaID'
def map_consumption_to_model_format(row):
    year = row['Fecha'][:4]
    quarter_num = int(row['Fecha'][5])
    quarter_map = {1: 'Q1', 2: 'Q2', 3: 'Q3', 4: 'Q4'}
    time_period = f"{year}{quarter_map[quarter_num]}"

    peru_consumption = row['Privado Perú']
    colombia_consumption = row['Privado Colombia']

    peru_row = {'refAreaID': 'PE', 'timePeriod': time_period, 'Consumption_it': peru_consumption}
    colombia_row = {'refAreaID': 'CO', 'timePeriod': time_period, 'Consumption_it': colombia_consumption}

    return pd.Series([peru_row, colombia_row])

# Apply the mapping function and expand the resulting Series of dictionaries into rows
df_consumption_processed = df_consumption.apply(map_consumption_to_model_format, axis=1).explode().apply(pd.Series)

# Merge df_final with the consumption data
df_merged = pd.merge(df_final, df_consumption_processed, on=['refAreaID', 'timePeriod'], how='left')

# Display the first few rows of the merged DataFrame
print(df_merged.head())
```
